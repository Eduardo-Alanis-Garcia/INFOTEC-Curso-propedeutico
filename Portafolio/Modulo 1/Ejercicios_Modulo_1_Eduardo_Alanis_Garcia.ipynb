2. Hacer un programa en Python que resuelva la siguiente ecuación de 2o grado.
El usuario ingresa los valores de a, b y c.

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$

Nota: Recordemos que una ecuacion de segundo grado es de la forma: $ax^2  + bx + c$

```python
import math

def resolver_cuadratica(a, b, c):

  try:
    a = float(a)
    b = float(b)
    c = float(c)
  except ValueError:
    print("Debes colocar valores numericos")
    return None


  r = b**2 - 4*a*c            # Lo defini como "r" por dentro de la raiz

  if r > 0:                  # Soluciones reales
    x1 = (-b + math.sqrt(r))/(2*a)
    x2 = (-b - math.sqrt(r))/(2*a)
    return f"Dos soluciones reales: x1 = {x1}, x2 = {x2}"

  elif r == 0:               # Solucion Unica
    x = -b / (2*a)
    return f"Una solución real: x = {x}"

  else:
    real = -b/(2*a)
    imaginaria = math.sqrt(-r)/(2*a)
    return f"Soluciones complejas: x1 = {real} + {imaginaria}i, x2 = {real} - {imaginaria}i"
```

```python
print(f"La ecuacion cuadratica es de la forma ax^2 + bx + c")
a = input("Ingresa a: ")
b = input("Ingresa b: ")
c = input("Ingresa c: ")
print(f"Esta ecuación es de la forma ({a})x^2 + ({b})x + ({c})")
print(resolver_cuadratica(a=a, b=b, c=c))
```

3. Hacer un programa en Python que calcule el salario semanal de un obrero,
según los siguientes criterios:
Si trabajó 40 horas o menos, sele paga $16 por hora.

*   Si trabajó 40 horas o menos, sele paga \$16 por hora.
*   Si trabajó más de 40 horas, se le paga \$16 por las primeras 40 horas, y \$20
por cada hora extra.

```python
def salario_semanal(horas):

  try:
    horas = float(horas)
  except ValueError:
    print("Debes colocar valores numericos")
    return None


  if horas <= 0:
    print("Colocar horas con sentido, debe ser valores positivos")
    return None

  elif horas <= 40:
    salario = horas * 16
  else:
    salario = (40 * 16) + ((horas - 40)*36)

  print("Tu salario semanal es de", salario, "pesos")
  return salario
```

4. Hacer un programa en Python que clasifique el tipo de triángulo a partir de sus lados ingresados:

| Triángulo  | Medidas de los lados                          |
|------------|----------------------------------------------|
| Equilátero | Los 3 son iguales                            |
| Isósceles  | Tiene 2 lados iguales y un lado distinto     |
| Escaleno   | Los 3 lados son distintos                    |

```python
def clasificar_triangulo(a, b, c):
  try:
    a = float(a)
    b = float(b)
    c = float(c)
  except ValueError:
    print("Debes colocar valores numericos")
    return None


  if a <= 0 or b <= 0 or c <= 0:
    print("Coloca valores positivos")
    return None

  # Debemos ver el caso que no satisface la desigualdad triangular
  if a + b <= c or a + c <= b or b + c <= a:
    print("No forman un triángulo válido")
    return None


  if a == b == c:
    print("Es un triángulo equilátero")
    return "Equilátero"

  elif a == b or a == c or b == c:
    print("Es un triángulo isósceles")
    return "Isósceles"

  else:
    print("Es un triángulo escaleno")
    return "Escaleno"
```

```python
lado1 = input("Ingresa el lado a: ")
lado2 = input("Ingresa el lado b: ")
lado3 = input("Ingresa el lado c: ")

clasificar_triangulo(a= lado1, b= lado2, c= lado3)
```

5. Hacer un programa en Python que calcule el Índice de Masa Corporal de una
persona (IMC = peso[kg] / altura [m]) e indique el estado en el que se encuentra
esa persona en función del valor de IMC.

| Valor de IMC | Diagnostico                                      |
|-----------|--------------------------------------------------|
| < 16      | Criterio de ingreso en hospital                  |
| 16 - 17   | Infrapeso                                       |
| 17 - 18   | Bajo peso                                       |
| 18 - 25   | Peso normal (saludable)                          |
| 25 - 30   | Sobrepeso (obesidad de grado I)                  |
| 30 - 35   | Sobrepeso crónico (obesidad de grado II)         |
| 35 - 40   | Obesidad premórbida (obesidad de grado III)      |
| > 40      | Obesidad mórbida (obesidad de grado IV)          |


```python
def calcular_imc(peso, altura):

  try:
    peso = float(peso)
    altura = float(altura)
  except ValueError:
    print("Debes colocar valores numericos")
    return None

  if peso <= 0 or altura <= 0:
    print("Coloca valores positivos para el peso y altura")
    return None

  imc = peso/(altura**2)


  if imc < 16:
    diagnostico = "Criterio de ingreso en hospital"
  elif 16 <= imc < 17:
    diagnostico = "Infrapeso"
  elif 17 <= imc < 18:
    diagnostico = "Bajo peso"
  elif 18 <= imc < 25:
    diagnostico = "Peso normal (saludable)"
  elif 25 <= imc < 30:
    diagnostico = "Sobrepeso (obesidad de grado I)"
  elif 30 <= imc < 35:
    diagnostico = "Sobrepeso crónico (obesidad de grado II)"
  elif 35 <= imc < 40:
    diagnostico = "Obesidad premórbida (obesidad de grado III)"
  else:
    diagnostico = "Obesidad mórbida (obesidad de grado IV)"

  print("IMC es: ", imc)
  print("Diagnóstico: ", diagnostico)

  return imc, diagnostico
```

```python
altura_usuario = input("Ingrese su altura en metros: ")
peso_usuario = input("Ingrese su peso: ")
calcular_imc(altura=altura_usuario, peso=peso_usuario)
```

muestre la lista de números pares inferiores a él.

```python
def numeros_pares_inferiores(n):

  try:
    n = int(n)
  except ValueError:
    print("Debes colocar un numero entero")
    return None

  # Considere los positivos unicamente, para poder comenzar el for con un valor mayor que 0 sin problemas
  if n <= 0:
    print("El número debe ser positivo")
    return None

  pares = []

  for i in range(0, n):    # Cpnsideramos solo numeros mayores iguales que 0, porque podemos poner -1000 y sigue satisfaciendo el ejercicio. Por eso mejor solo >=0
    if i % 2 == 0:
      pares.append(i)

  print("El numero entero que colocaste es", n, "\n", "los numeros pares inferiores a el son: ", pares, "\n", "en total son: ", len(pares))
  return pares
```

11. Hacer un programa en Python que permita registrar las calificaciones de varios estudiantes. Cada estudiante tiene un nombre y una lista de calificaciones, implementa funciones para:


- a) Agregar un estudiante  
- b) Agregar una calificación  
- c) Calcular el promedio de cada estudiante  



Ejemplo de la estructura de datos

```python
estudiantes = {
    "Pedro": [8.5, 9.0],
    "Maria": [7.0, 9.5, 8.0]
}
```

```python
def agregar_estudiante(estudiantes, nombre):
  if nombre in estudiantes:
    print("El estudiante ya existe")
  else:
    estudiantes[nombre] = []
    print("Estudiante" , nombre, "agregado")
```

```python
agregar_estudiante(estudiantes=estudiantes, nombre="Juan")
estudiantes
```

```python
def agregar_calificacion(estudiantes, nombre, calificacion):
  if nombre not in estudiantes:
    print("El estudiante no existe")
    return

  try:
    calificacion = float(calificacion)
  except ValueError:
    print("Debes colocar valores numericos")
    return

  estudiantes[nombre].append(calificacion)
  print("Calificación agregada a", nombre)
```

```python
agregar_calificacion(estudiantes, "Juan", 8)
agregar_calificacion(estudiantes, "Juan", 9.5)
estudiantes
```

```python
def calcular_promedio(estudiantes, nombre):
  if nombre not in estudiantes:
    print("El estudiante no existe")
    return None

  if len(estudiantes[nombre]) == 0:
    print("Agregale una calificacion al estudiante")
    return None

  promedio = sum(estudiantes[nombre]) / len(estudiantes[nombre])
  print("Promedio de", nombre, "es de", promedio)
  return promedio
```

```python
calcular_promedio(estudiantes, "Pedro")
calcular_promedio(estudiantes, "Juan")
calcular_promedio(estudiantes, "Maria")
estudiantes
```

12. Crea un programa en Python para gestionar un pequeño catálogo de

Cada producto tendrá:

- a) Un código clave

Implementa funciones para:

- a) Agregar productos  
- b) Consultar un producto por código  
- c) Calcular el valor total del inventario  

```python
productos = {
    "A001" : ("Camisa", 250.0, 10),
    "B002" : ("Pantalon", 400.0, 5)
}
```

```python
def agregar_producto(catalogo, codigo, nombre, precio, cantidad):
  if codigo in catalogo:
    print("El producto ya existe")
    return

  try:
    precio = float(precio)
    cantidad = int(cantidad)
  except ValueError:
    print("Debes colocar numeros, donde la cantidad debe ser entero")
    return

  if precio < 0 or cantidad < 0:
    print("Debemos considerar ambos valores positivos")
    return

  print("El producto", nombre, "fue agregado")
  catalogo[codigo] = (nombre, precio, cantidad)
```

```python
agregar_producto(catalogo=productos, codigo="A111", nombre="Playera", precio=150, cantidad=17)
```

```python
def consultar_producto(catalogo, codigo):
  if codigo not in catalogo:
    print("Producto no encontrado")
    return None



  nombre, precio, cantidad = catalogo[codigo]

  print("Codigo", codigo, "\n", "Nombre:", nombre, "\n" ,"Precio", precio, "\n" ,"Cantidad", cantidad)

  return catalogo[codigo]
```

```python
consultar_producto(catalogo=productos, codigo="A001")
```

```python
def valor_total_inventario(catalogo):
  total = 0

  for producto in catalogo.values():
    nombre, precio, cantidad = producto
    total = total + precio * cantidad

  print("Valor total del inventario:", total)
  return total
```

```python
valor_total_inventario(catalogo=productos)
```

13. Escribe una función en Python que reciba una frase y devuelva un diccionario
con la cantidad de veces que aparece cada palabra. Ignora mayúsculas y
signos de puntuación


```python
import string

def contar_palabras(frase):

  frase = frase.lower() #Habia metido el lower dentro del for, pero es mas efectivo asi dado que se aplica a toda la frase

  # Tuve que investigar esta funcion, dado que solo habia trabajado con String de forma muy detallada en R
  for signo in string.punctuation:
    frase = frase.replace(signo, "")

  palabras = frase.split()

  conteo = {}
  for palabra in palabras:
    conteo[palabra] = conteo.get(palabra, 0) + 1

  return conteo
```

```python
entrada = "Hola mundo hola mundo hola hola adios"
salida = contar_palabras(frase=entrada)
print(salida)
```

15. Crea una función en Python que reciba una lista de números y devuelva una tupla con los siguientes datos:

- a) número mayor  
- b) número menor
- c) promedio
- d) cantidad de números pares
- e) cantidad de números impares  

```python
def lista_numeros_a_dupla_ordenada(lista):
  mayor = max(lista)
  menor = min(lista)
  promedio = sum(lista)/len(lista)

  pares = 0
  impares = 0

  for num in lista:
    if num % 2 == 0:
        pares = pares + 1
    else:
        impares = impares + 1

  return (mayor, menor, promedio, pares, impares)
```

```python
numeros = [2, 5, 8, 3, 10]
lista_numeros_a_dupla_ordenada(lista=numeros)
```

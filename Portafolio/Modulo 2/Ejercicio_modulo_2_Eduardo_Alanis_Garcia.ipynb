## 5. Dividir los Datos en Entrenamiento y Prueba:

- Definir las características (X) y la variable objetivo (y), que es returned_next_month.
- Dividir los datos en un conjunto de entrenamiento y otro de prueba (80/20).

```python
from sklearn.model_selection import train_test_split
X = datos.drop(columns=["customer_id", "returned_next_month"])
y = datos["returned_next_month"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```

Las características de X son todas las variables, excepto "customer_id", ya que esta no aporta información útil para el modelo y solo funciona como un identificador único. También se excluye la variable objetivo. Ademas indicamos la division de los datos en un 80% para entrenamiento y 20% para prueba, donde indicamos random_state = 42 como semilla predeterminada (asi lo suelen hacer en bastantes ejemplos para hacer la misma reproduccion del modelo).

Nota: Antes de entrenar el modelo, es una buena práctica dividir el conjunto de entrenamiento en subconjuntos de entrenamiento y validación. Esto nos permite evaluar el rendimiento del modelo antes de hacer predicciones en el conjunto de prueba.

## 6. Entrenamiento de Modelos:

- Seleccionar tres modelos de clasificación: Regresión Logística, Árbol de Decisión y Bosque Aleatorio.
- Entrenar cada modelo usando los datos de entrenamiento.

```python
# Añadimos random_state=42 para tener siempre los mismo resultados, lo equivalente a una semilla

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Modelo
modelo_lr  = LogisticRegression(random_state=42)
modelo_dt  = DecisionTreeClassifier(random_state=42)
modelo_rf  = RandomForestClassifier(random_state=42)

# Entrenamiento
modelo_lr.fit(X_train, y_train)
modelo_dt.fit(X_train, y_train)
modelo_rf.fit(X_train, y_train)
```

Primero vamos a conocer un poco sobre los modelos:

| Modelo | Justificación |
|---|---|
| **Regresión Logística** | Es un modelo simple pero efectivo para problemas de clasificación binaria; útil cuando las relaciones entre variables son aproximadamente lineales |
| **Árbol de Decisión** | Captura relaciones no lineales; útil para tareas de clasificación, donde el objetivo es predecir una etiqueta o categoría |
| **Bosque Aleatorio** |  Ideal para problemas de clasificación, que utiliza una técnica de ensamble (bagging) para combinar múltiples árboles de decisión, el más robusto de los tres |


Como el objetivo es predecir si un cliente regresará el siguiente mes, cualquiera de los tres modelos puede funcionar bien:



*   La Regresión Logística es adecuada porque el problema tiene solo dos posibles respuestas: el cliente regresa o no regresa.
*   El Árbol de Decisión también sirve porque busca clasificar al cliente en una categoría según su comportamiento.
*   El Bosque Aleatorio es útil porque mejora la precisión al combinar varios árboles de decisión, haciendo el modelo más estable y confiable.





## 7. Evaluación Inicial de Modelos:

- Evaluar cada modelo en el conjunto de prueba usando métricas de rendimiento: exactitud, precisión, recall, F1 y AUC-ROC.
- Comparar los resultados y documentarlos para seleccionar el mejor modelo.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, roc_curve
import matplotlib.pyplot as plt

modelos = {
    "Regresión Logística": modelo_lr,
    "Árbol de Decisión":   modelo_dt,
    "Bosque Aleatorio":    modelo_rf,
}

resultados_modelos = pd.DataFrame(columns=["nombre", "exactitud", "precision", "recall", "f1", "auc_roc"])

plt.figure(figsize=(8, 6))

for nombre, modelo in modelos.items():

    y_pred = modelo.predict(X_test)
    y_prob = modelo.predict_proba(X_test)[:, 1]

    exactitud = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall    = recall_score(y_test, y_pred)
    f1        = f1_score(y_test, y_pred)
    auc_roc   = roc_auc_score(y_test, y_prob)

    interes = pd.DataFrame([{
        "nombre":    nombre,
        "exactitud": exactitud,
        "precision": precision,
        "recall":    recall,
        "f1":        f1,
        "auc_roc":   auc_roc,
    }])
    resultados_modelos = pd.concat([resultados_modelos, interes], ignore_index=True)

    # Curva ROC individual
    print("\n", end="\n\n")
    print(f"{nombre}")
    print(classification_report(y_test, y_pred, target_names=["No regresa", "Regresa"]))

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{nombre} (AUC = {auc_roc:.2f})")


plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Clasificador aleatorio") # Línea de referencia
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curvas ROC")
plt.legend()
plt.grid()

plt.show()

print("\n", end= "\n\n\n")
print(resultados_modelos)
```

### Métrica ROC AUC

La métrica **ROC AUC** (*Receiver Operating Characteristic - Area Under the Curve*) es una medida comúnmente utilizada para evaluar el desempeño de un modelo de clasificación, especialmente en problemas de clasificación binaria.

Combina dos aspectos clave:

#### 1. Curva ROC

Representa gráficamente la relación entre la tasa de verdaderos positivos (*True Positive Rate* o *Sensitivity*) y la tasa de falsos positivos (*False Positive Rate*) a diferentes umbrales de decisión.

Esto permite analizar cómo varía la capacidad del modelo para distinguir entre las clases positivas y negativas según se ajusta el umbral.


#### 2. AUC (Area Under the Curve)

Es el área bajo la curva ROC, que cuantifica la habilidad del modelo para clasificar correctamente.

Su valor está entre 0 y 1:

- Un **AUC = 1** indica un modelo perfecto.
- Un **AUC = 0.5** implica que el modelo no es mejor que adivinar al azar.
- Un **AUC cercano a 0** significa que el modelo clasifica de manera opuesta a lo esperado.

En resumen, ROC AUC es útil para comparar modelos y medir su capacidad de discriminar entre clases, sin depender de un único umbral de decisión.

#### 1. Tasa de Verdaderos Positivos (TPR) — Eje vertical

- También conocida como **sensibilidad** o **recall**, mide la proporción de positivos reales que el modelo clasifica correctamente.
- Fórmula:

$$
TPR = \frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos} + \text{Falsos Negativos}}
$$



#### 2. Tasa de Falsos Positivos (FPR) — Eje horizontal

- Mide la proporción de negativos reales que el modelo clasifica incorrectamente como positivos.
- Fórmula:

$$
FPR = \frac{\text{Falsos Positivos}}{\text{Falsos Positivos} + \text{Verdaderos Negativos}}
$$



#### 3. Línea diagonal

- La línea punteada representa un clasificador aleatorio, donde la TPR es igual a la FPR para todos los umbrales (como lanzar una moneda).
- Un buen modelo tendrá su curva ROC por encima de esta línea.



#### 4. Curva ROC

- La curva azul muestra el desempeño del modelo en distintos umbrales de decisión.
- Una curva más pronunciada (más cercana a la esquina superior izquierda) indica un mejor modelo, porque logra una alta TPR con una baja FPR.

## 8. Optimización del Mejor Modelo:

- Realizar optimización de hiperparámetros (Grid Search) en el modelo seleccionado (en este caso, Bosque Aleatorio).

Tenemos que la documentacion oficial de RandomForestClassifier tiene las siguientes parametros:

```python
"""
Lo siguiente me estoy basando con:

Documentacion oficial: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html
Ejemplo: https://www.kaggle.com/code/sociopath00/random-forest-using-gridsearchcv

"""

from sklearn.model_selection import GridSearchCV

# Cuadrícula de hiperparámetros a explorar
param_grid = {
    'n_estimators':      [50, 100, 200],
    'max_depth':         [None, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf':  [1, 2]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Mejores hiperparámetros encontrados: \n", grid_search.best_params_)
```

```python
modelo_rf_optimo = RandomForestClassifier(random_state=42, max_depth=None, min_samples_leaf=1, min_samples_split=2, n_estimators=50)
modelo_rf_optimo.fit(X_train, y_train)
```

Los hiperparámetros explorados son:

| Hiperparámetro | Descripción | Valores explorados |
|---|---|---|
| `n_estimators` | Número de árboles | 50, 100, 200 |
| `max_depth` | Profundidad máxima de cada árbol | None, 5, 10 |
| `min_samples_split` | Mínimo de muestras para dividir un nodo | 2, 5 |
| `min_samples_leaf` | Mínimo de muestras en una hoja | 1, 2 |


Nota: Usamos estos porque realizabamos la prueba con bastantes y me sacaba error, ademas que el codigo tardaba bastante.


Tenemos que el modelo usa 50 árboles (n_estimators = 50), lo cual no es una cantidad muy grande comparado con las otras opciones que se probaron, que eran 100 y 200 árboles.

Además, max_depth = None significa que los árboles pueden crecer sin límite.

Por otro lado, min_samples_split = 2 indica que un nodo puede dividirse con apenas 2 muestras, y min_samples_leaf = 1 significa que cada hoja puede contener solo una muestra.

Por todo lo anterior diria que los parametros hacen que el modelo sea bastante flexible.
